#### Using CPU

```python
import torch
device = torch.device('cpu')
x = torch.rand((10000, 10000), dtype=torch.float32)
y = torch.rand((10000, 10000), dtype=torch.float32)
x = x.to(device)
y = y.to(device)
```

```python
%%timeit
x * y
```

#### Using NVIDIA GPU

```python
import torch
device = torch.device('cuda')
x = torch.rand((10000, 10000), dtype=torch.float32)
y = torch.rand((10000, 10000), dtype=torch.float32)
x = x.to(device)
y = y.to(device)
```

```python
%%timeit
x * y
```
